```python
import os
import pandas as pd
import io

# 현재 작업 디렉토리 확인 및 출력
current_directory = os.getcwd()
print(f"현재 작업 디렉토리: {current_directory}")

# AllFile 폴더의 경로 설정
allfile_directory = os.path.join(current_directory, "AllFile")

# AllFile 폴더의 파일 목록 가져오기
csv_files = [file for file in os.listdir(allfile_directory) if file.endswith('.csv')]

# 여러 CSV 파일을 하나의 DataFrame으로 합치기
combined_df = pd.DataFrame()

for file in csv_files:
    file_path = os.path.join(allfile_directory, file)
    
    try:
        # CSV 파일 읽기, 주석(#)으로 시작하는 줄은 건너뜀
        with open(file_path, 'r', encoding='utf-8-sig') as f:
            lines = f.readlines()
        
        # 주석(#)으로 시작하지 않는 데이터만 필터링
        non_comment_lines = [line for line in lines if not line.startswith('#')]
        
        # 데이터프레임으로 변환
        df = pd.read_csv(io.StringIO(''.join(non_comment_lines)), on_bad_lines='skip')
        
        # 데이터프레임 합치기
        combined_df = pd.concat([combined_df, df], ignore_index=True)
        
        print(f"파일 '{file}' 처리 성공")
    
    except pd.errors.ParserError as e:
        print(f"파일 '{file}'에서 문제가 발생했습니다: {e}")
    except Exception as e:
        print(f"파일 '{file}' 처리 중 일반 오류 발생: {e}")

# 합쳐진 CSV 파일 저장
output_file_path = os.path.join(allfile_directory, "뉴스인사정책.csv")
combined_df.to_csv(output_file_path, index=False, encoding='utf-8-sig')

print(f"모든 CSV 파일이 '{output_file_path}'로 성공적으로 합쳐졌습니다.")
```
